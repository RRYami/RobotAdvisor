```python
import plotly.graph_objects as go
import pandas as pd
import numpy as np
import yfinance as yf
```

```python
nvda = pd.read_csv(r"C:\Users\renar\PythonRobotAdv\data\transformed\Adj_NVDA_test.csv")
nvda.set_index('Unnamed: 0', inplace=True)
nvda.index = pd.to_datetime(nvda.index)
```

```python
yf_nvda = yf.download('NVDA', start='1999-11-01', end='2024-08-21')['Adj Close']
```

```python
df = pd.concat([nvda, yf_nvda], axis=1, join='inner')
df['Difference'] = df['Adj Close'] - df['Close']
```

```python
np.std(df['Difference'])
```

```python
fig = go.Figure()
# fig.add_trace(go.Scatter(x=df.index, y=df['Close'], mode='lines', name='Close'))
# fig.add_trace(go.Scatter(x=df.index, y=df['Adj Close'], mode='lines', name='Adj Close'))
fig.add_trace(go.Scatter(x=df.index, y=df['Difference'], mode='lines', name='Difference'))
fig.show()
```
